```python
#use numpy, pandas package
import numpy as np 
import pandas as pd

#read the csv file 
data1 = pd.read_csv('6ef495654a3afcaa.csv')
data1
```

```python
#drop the meaningless column
data2 = data1.drop(['gvkey','datadate','fyearq','fqtr','fyr'],axis =1)
data2
```

```python
#first try of using describe function
result = data2['xrdq'].describe()
pd.DataFrame(result)
```

```python
#just select column which dtype is "number"
data2.describe(include=['number'])
```

```python
#to count how mnay percentage of missing data for each column
data2.isnull().mean()
```

```python
#use the column which has less than 50% missing data
data3 = data2.loc[:, data1.isnull().mean() < .5]
data4 = data3.describe(include=['number'])
data4
```

```python
#for counting the missing
data3.isnull().sum()
data_final = data4.append(data3.isnull().sum(), ignore_index=True)
data_final.rename(index={0:'observations',1:'mean',2:'std',3:'min',4:'p25',5:'Median',6:'P75',7:'Max',8:'Missing'}, inplace=True)
data_final
```
